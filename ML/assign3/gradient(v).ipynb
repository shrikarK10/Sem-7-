```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x + 3)**2

def grad_f(x):
    return 2 * (x + 3)
```

```python
def gradient_descent(starting_point, learning_rate, num_iterations):
    x = starting_point
    x_history = [x] 
    
    for i in range(num_iterations):
        x = x - learning_rate * grad_f(x)
        x_history.append(x)
        
    return x, x_history
```

```python
starting_point = 2.0  
learning_rate = 0.1
num_iterations = 500 
```

```python
final_x, x_history = gradient_descent(starting_point, learning_rate, num_iterations)
```

```python
print(f"Local minima found at x = {final_x}")
```

```python
x_values = np.linspace(-10, 4, 400)
y_values = f(x_values)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label="f(x) = (x + 3)^2", color="cyan")
plt.scatter(x_history, [f(x) for x in x_history], color="black", label="Iterations", zorder=5)
plt.plot(x_history, [f(x) for x in x_history], color="black", linestyle="--", alpha=0.6)
plt.title("Gradient Descent Optimization")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
